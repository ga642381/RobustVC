```python
import augment
import torchaudio
import librosa
import torch
import IPython.display as ipd
```

```python
wav = torchaudio.load("/fortress/dataset/VCTK-Corpus/wav48/p225/p225_001.wav")
```

```python
wav, sr = librosa.load("/fortress/dataset/VCTK-Corpus/wav48/p225/p225_001.wav", 16000)
```

```python
ipd.Audio(wav, rate=sr)
```

```python
wav_tensor = torch.tensor(wav)
```

```python
wav_tensor
```

# Wav Aug

```python
wavaug_chain = augment.EffectChain()
wavaug_chain = wavaug_chain.sinc("-a", "120", "250-100")
noisy_wav = wavaug_chain.apply(wav_tensor, src_info={"rate": sr})
```

```python
ipd.Audio(noisy_wav.cpu().numpy(), rate=sr)
```
